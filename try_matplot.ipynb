```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.show()
```

```python
p_x = np.random.rand(100)
p_y = np.random.rand(100)
plt.scatter(p_x, p_y)
plt.show()
```

```python
data = np.random.randn(100000)
plt.hist(data)
plt.show()
```
